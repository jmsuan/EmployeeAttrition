### 2. interactive()

```python
from IPython.display import display

def formula(a, b, c):
    print(a*5 + b*7 + c*10)

gui = interactive(formula, a=1, b=1, c=fixed(10))

display(gui)
```

```python
import ipywidgets as widgets
from IPython.display import display

slider = widgets.IntSlider(description='Value:', min=0, max=100)
bar = widgets.HTML(value=None, layout=widgets.Layout(width='0%', height='30px', background_color='lightblue'))
bar_label = widgets.Label(value='0%')

def update_bar(change):
    value = change['new']
    bar.value = value
    bar.layout.width = f'{value}%'
    bar_label.value = f'{value}%'

slider.observe(update_bar, names='value')

display(slider, widgets.HBox([bar, bar_label]))
```

```python
import ipywidgets as widgets
from IPython.display import display

slider = widgets.IntSlider(description='Value:', min=0, max=100)
bar = widgets.HTML(value='', layout=widgets.Layout(width='0.3%', height='60px', background='red'))
bar_label = widgets.Label(value='0% likely to leave the company')
bar.value = f'<div style="background:red; height: 60px; width: 100%"></div>'

def update_bar(change):
    value = change['new']
    bar.layout.width = f'{value * 0.8 + 0.3}%'
    bar_label.value = f'{value}% likely to leave the company'

slider.observe(update_bar, names='value')

display(widgets.VBox([slider, widgets.HBox([bar, bar_label])]))
```
